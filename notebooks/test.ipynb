```python
import pandas as pd
import requests
from sqlalchemy import create_engine
import os

try:
    df_csv = pd.read_csv("student-mat 2.csv")
    print("CSV الطلاب: ", df_csv.shape)
except Exception:
    print(" ملف CSV الطلاب غير موجود، سيتم تجاهله!")
    df_csv = pd.DataFrame()

try:
    df_excel = pd.read_excel("gestion.xlsx")
    print("Excel البيانات: ", df_excel.shape)
except Exception:
    print(" ملف Excel غير موجود أو خطأ في القراءة، سيتم تجاهله!")
    df_excel = pd.DataFrame()

try:
    df_movies = pd.read_csv("movies_metadata.csv", low_memory=False)
    df_movies.to_json("movies_metadata.json", orient="records")
    print("Movies CSV تم تحويله إلى JSON: movies_metadata.json")
except Exception:
    print(" ملف Movies CSV غير موجود أو خطأ في القراءة، سيتم تجاهله!")
    df_movies = pd.DataFrame()

try:
    engine = create_engine("mysql+pymysql://root:password@localhost/school")
    df_sql = pd.read_sql("SELECT * FROM etudiants", engine)
    print("SQL البيانات: ", df_sql.shape)
except Exception:
    print(" فشل الاتصال ب MySQL، سيتم تجاهلها!")
    df_sql = pd.DataFrame()

try:
    response = requests.get("https://jsonplaceholder.typicode.com/posts", timeout=10)
    response.raise_for_status()
    df_api = pd.DataFrame(response.json())
    print("API البيانات: ", df_api.shape)
except Exception:
    print(" فشل جلب بيانات API، سيتم تجاهله!")
    df_api = pd.DataFrame()

dfs = [df_csv, df_excel, df_movies, df_sql, df_api]
dfs_to_merge = [df for df in dfs if not df.empty]

if dfs_to_merge:
    df_final = pd.concat(dfs_to_merge, ignore_index=True, sort=False)
    print(" البيانات النهائية: ", df_final.shape)
    df_final.to_csv("final_dataset.csv", index=False)
    print(" تم حفظ الملف النهائي: final_dataset.csv")
else:
    print(" لا توجد بيانات للدمج!")
```
